# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000); #gen random 1000 sample  
print(len(x),type(x))
sns.set_style('whitegrid')
sns.kdeplot(x, bw_method=0.5)
#kernel density estimation
plt.show()
```

```python
stats.kstest(x, 'norm')
#KS test compare x with normal distribution
#as high p value hence random variable x is normallly ditributed 
```

# NEGATIVE CASE

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw_method=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
#p value is 0 therefore y is not normally ditributed
```
