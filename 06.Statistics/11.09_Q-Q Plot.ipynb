# Q-Q Plot FROM NORMAL DISTRIBUTION

```python
#Q-Q plot from Copy
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)--> Standard Normal Distribution
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)
#loc = mean , scale = standard deviation ,size=generate 1000 samples

# 0 to 100th percentiles of std-normal
for i in range(0,101):
     print(i, np.percentile(std_normal,i))
#print("25th percentile of arr : ", np.percentile(arr, 25))
#what i want here is all percentiles 
```

```python
import matplotlib.pyplot as plt
```

```python
# generate 100 samples from N(20,5)
#mean=20 and SD=5
measurements = np.random.normal(loc = 20, scale = 5, size=100) 
#use 1000

stats.probplot(measurements, dist="norm", plot=pylab)
#compare with normal std dist
#propplot--> Q-Q Plot
pylab.show()
 
```

```python
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=1000) 
#try size=1000
#plt.xlim(-4, 4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
# generate 30 sanples from N(20,5)
measurements = np.random.normal(loc = 0, scale = 1, size=30) 
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

# Q-Q Plot FROM UNIFORM DISTRIBUTION

```python
# generate 100 samples from N(20,5)
#mean=20 and SD=5
measurements = np.random.uniform(low =-1, high= 1, size=100) 
#use 1000

stats.probplot(measurements, dist="norm", plot=pylab)
#compare with normal std dist
pylab.show()
```
