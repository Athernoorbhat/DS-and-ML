# PYTHON Program to Print HCF of two number

```python
def compute_HCF(a,b):
    """
    Computing the HCF of Two Numbers
    """
    smaller = b if a>b else a
    
    hcf=1
    for i in range(1,smaller+1):
        if (a % i == 0) and (b % i == 0):
            hcf=i
    return hcf

num1=98
num2=78

print("HCF of {0} and {1} is : {2}".format(num1,num2,compute_HCF(num1,num2)))
```

# PYTHON PROGRAM TO MAKE SIMPE CALCULATOR

```python
def add(a,b):
    """
    This function addition two numbers
    """
    return a+b

def multiply(a,b):
    """
    This function multiply two numbers
    """
    return a*b

def subtract (a,b):
    """
    This function subracts two numbers
    """
    return a-b

def division(a,b):
    """
    This function divides two numbers
    """
    return a/b

print("Select Options")
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")

#take inputfrom user
choice = int(input("Enter choice1/2/3/4:"))

num1 =float(input("Enter first number:"))
num2 =float(input("Enter Second number:"))

if choice==1:
    print("Addition of {0} and {1} is {2}".format(num1,num2,add(num1,num2)))

elif choice==2:
    print("Subraction of {0} and {1} is {2}".format(num1,num2,subract(num1,num2)))

elif choice==3:
    print("Multiplication of {0} and {1} is {2}".format(num1,num2,multiply(num1,num2)))

elif choice==4:
    print("Division of {0} and {1} is {2}".format(num1,num2,division(num1,num2)))
else:
    print("Invalid Choice")
    
```

# Different Forms of Arguments

# 1.Default Arguments

```python
def greet(name ,msg="Good Evening"):
    
    """
    This function greets to person with provided msg 
    however if msg is not provided , default 'Good Evening' 
    """
    print("Hello {0} , {1}".format(name,msg))

greet ("Ather", "Good Morning")
greet ("Ather")
    
```

```python
def greet(msg="Good Evening",name):
    print("Hello {0} , {1}".format(msg,name))
greet ("Ather", "Good Morning")
greet ("Ather")
```

# 2.Keyword Arguments

kwarg allows you to pass keyworded variable length of arguments to a function.You should use 
**kwargs if you want to handle named arguments in a function

```python
def greet(**kwargs):
    """
    This function greets to person with provided message
    """
    print(kwargs)
    
    if kwargs:
        print("Hello {0} , {1} ".format(kwargs['name'],kwargs['msg']))

#greet ("Ather", "Good Morning")
#here keyword and value unlike upper casses
greet(name="Ather",msg="Good Evening")
```

# 3.Arbitary Arguments

Sometimes you dont know in advance the no of arguments that will be passed into a function.So handle that situation through function calls with arbitary number of arguments.

```python
def greet(*args):
    """
    This function greets all persons in the name tuple
    """
    print(args) #tuple print
    
    for name in args:
        print("Hello, {0} ".format(name))

greet("Ather", "Azhar","Akhtar","Anzar")
```
