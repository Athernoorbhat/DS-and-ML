# Mean , Variance and Standard Deviation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load data set
iris=pd.read_csv("iris.csv")
iris
```

```python
#.loc() to crete subset on specific conditions

iris_Setosa = iris.loc[iris["variety"] == "Setosa"]
iris_Virginica = iris.loc[iris["variety"] == "Virginica"]
iris_Versicolor = iris.loc[iris["variety"] == "Versicolor"]
#print(iris_Setosa["sepal.length"])
```

# Mean

```python
print("Means:")
print(np.mean(iris_Setosa["petal.length"])) #Small petal legth
print(np.mean(iris_Virginica["petal.length"]))  #close to versicolor
print(np.mean(iris_Versicolor["petal.length"])) #close to virginica
print("-------------------------------------")
#Mean with outlier
print("Outlier:")
print(np.mean(np.append(iris_Setosa["petal.length"],50)));

#somebody entered wrong reading of 50cm petal length
```

# Standard Deviation

```python
print("Standard Deviations:")
print(np.std(iris_Setosa["petal.length"])) #lower SD
print(np.std(iris_Virginica["petal.length"])) #higer SD
print(np.std(iris_Versicolor["petal.length"])) #theek thak SD
```

# Variance

```python
print("Variance:")
print(np.var(iris_Setosa["petal.length"])) 
print(np.var(iris_Virginica["petal.length"])) 
print(np.var(iris_Versicolor["petal.length"])) 
```

Variance=(Standard Dev)^2

```python
print("Standard Deviations:")
print(np.std(iris_Setosa["petal.length"])**2) 
print(np.std(iris_Virginica["petal.length"])**2) 
print(np.std(iris_Versicolor["petal.length"])**2) 
```

# Median

```python
print("Meadian:")
print(np.median(iris_Setosa["petal.length"])) 
print(np.median(iris_Virginica["petal.length"])) #sim to mean
print(np.median(iris_Versicolor["petal.length"])) #similar to mena
print("-------------------------------------")
#Mean with outlier
print("Outlier:")
print(np.median(np.append(iris_Setosa["petal.length"],50))); #same as median

#Even with outlier the no is same 
#not corrupted by outlier
```

# Quantile

```python
print("Quantile:")
print(np.percentile(iris_Setosa["petal.length"],np.arange(0,100,25))) 
print(np.percentile(iris_Virginica["petal.length"],np.arange(0,100,25))) 
print(np.percentile(iris_Versicolor["petal.length"],np.arange(0,100,25)))
```

# 90th Percentile

```python
print("90th Quantile:")
print(np.percentile(iris_Setosa["petal.length"],90)) 
print(np.percentile(iris_Virginica["petal.length"],90)) 
print(np.percentile(iris_Versicolor["petal.length"],90))
```

# 25th Percentile

```python
print("25th Quantile:")
print(np.percentile(iris_Setosa["petal.length"],25)) 
print(np.percentile(iris_Virginica["petal.length"],25)) 
print(np.percentile(iris_Versicolor["petal.length"],25))
```

# Mad

```python
from statsmodels import robust
print("Median Absolute Deviation:")
print(robust.mad(iris_Setosa["petal.length"])) 
print(robust.mad(iris_Virginica["petal.length"])) 
print(robust.mad(iris_Versicolor["petal.length"]))
```

# IQR

```python
# Calculate the interquartile range for iris_Setosa
q1_Setosa = np.percentile(iris_Setosa["petal.length"], 25)
q3_Setosa = np.percentile(iris_Setosa["petal.length"], 75)
iqr_Setosa = q3_Setosa - q1_Setosa
print("IQR for iris_Setosa:", iqr_Setosa)

# Calculate the interquartile range for iris_Virginica
q1_Virginica = np.percentile(iris_Virginica["petal.length"], 25)
q3_Virginica = np.percentile(iris_Virginica["petal.length"], 75)
iqr_Virginica = q3_Virginica - q1_Virginica
print("IQR for iris_Virginica:", iqr_Virginica)

# Calculate the interquartile range for iris_Versicolor
q1_Versicolor = np.percentile(iris_Versicolor["petal.length"], 25)
q3_Versicolor = np.percentile(iris_Versicolor["petal.length"], 75)
iqr_Versicolor = q3_Versicolor - q1_Versicolor
print("IQR for iris_Versicolor:", iqr_Versicolor)
```
