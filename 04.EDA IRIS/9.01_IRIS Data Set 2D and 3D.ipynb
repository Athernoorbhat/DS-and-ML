# 2D Scatter Plot

```python
#2-D scatter plot
#always understand the axis:labels and scale.
iris.plot(kind='scatter',x='sepal.length',y='sepal.width')

plt.show()
#plot(kind='bar')
#plot(kind='hist')

#no much from it
#what if we color code
#start point is not (0,0)
#see the range of x-axis and y-axis
```

# Observations
1.not much sense as such

2.starting point is not(0,0) eather approx (4,1)

```python
#2-D scatter plot with color code
#sns --> seaborn
sns.set_style("whitegrid") #gives write grid structure
sns.FacetGrid(iris, hue="variety" ,height=5) \
    .map(plt.scatter,'sepal.length','sepal.width')\
    .add_legend();
#hue=variety--> means color it on variety height kitne graph ki
#.map(plt.scatter,'x=sepal.length','y=sepal.width') 
#legend--> gives names

plt.show()
```

# Observations
1.Sepal length and sepal width feature can distinguish setosa flower from other
2.Sperating Versicolor and virginica is much harder as they are overlapping

# 3-D Scatter Plot

https://plotly.com/python/3d-scatter-plots/

1.Cannot visualize higer dimensional pattern in 3D and 4D.

2.Only possible to view 2D pattern, simple trick of pairplot comes to help

```python
#pairwise scatter plot: Pair-Plot
plt.close();
sns.set_style("whitegrid") #gives write grid structure
sns.pairplot(iris, hue="variety" ,height=5);
plt.show()
```

# Multivariate Probability Density (MPD) ,Contour Plot

```python
iris_Setosa = iris.loc[iris["variety"] == "Setosa"]
iris_Virginica = iris.loc[iris["variety"] == "Virginica"]
iris_Versicolor = iris.loc[iris["variety"] == "Versicolor"]
```

```python
#Density Plot,contour plot
sns.jointplot(data=iris_Setosa,kind='kde',x='petal.length',y='petal.width',fill=True, thresh=0, cmap='Blues')
plt.show();

#'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
```

```python
#3D  Density Plot,contour plot
sns.jointplot(data=iris,hue='variety',kind='kde',x='petal.length',y='petal.width')
plt.show();
```
