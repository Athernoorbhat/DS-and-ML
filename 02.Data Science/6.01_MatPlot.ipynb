# PyPlot 

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,10,1])
#no x value so(0,2),(1,4),(2,10),(3,1)
plt.ylabel("Numbers") #yaxis
plt.xlabel("Indices") #xaxis
plt.title("MyPlot")   #title of plot
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
#(1,1),(2,4),(3,9),(4,16)
plt.ylabel("Squraes") #yaxis
plt.xlabel("Numbers") #xaxis
plt.title("Plot")     #title of plot
plt.grid()            #grid on

plt.show()            
```

```python
plt.plot([1,2,4,6],[1,4,16,36],'ro') #red color and o-dot 

plt.grid()            #grid on

plt.show()            
```

```python
import numpy as np
t=np.arange(0.,5.,0.2)

plt.plot(t,t**2,'b--',label='^2') #blue dashed line
plt.plot(t,t**3,'rs', label='^3')  #red with squares
plt.plot(t,t**4,'g^', label='^4')  #green with ^
plt.grid()
plt.legend() #which line corresponds to what upper square in left
plt.show()
```

# Controlling Line Properties

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=1.0)
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)
#line[0]=x1,y1
#line[1]=x2,y2

#use keywordargs
plt.setp(lines[0],color='r',linewidth=2.0) #set properties

#or use matlab style string value pairs
plt.setp(lines[1],'color','g','linewidth',2.0) #set properties

plt.grid()
```

# Working With Multiple Figures and axes

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1) #creates figure --> empty

#draw subplots in the figure
#subplot() command specifies numrows,numcols.
#fignum where fignum ranges from 1 to numrows*numcols

plt.subplot(211) #2 rows and 1 colums and 1 fig no
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(212) # 2 rows ,1 colm , 2 fig 
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)               #the first figure

plt.subplot(211)            #2 subplots ,1 row, 1st fig
plt.plot([1,2,3])              

plt.subplot(212)            #2 subplots ,1 row, 2nd fig
plt.plot([4,5,6])

plt.figure(2)               #the 2nd figure

plt.plot([4,5,6])           #creates subplot(111) by default

plt.figure(1)               # go back to fig 1 again
plt.subplot(211)
plt.title('Easy')
plt.show()
```
